broadcasting is performing operations on array with different shapes
rules:
dimensions need to be the same size or
one of the dimensions need to be 1

```python
array1 = np.array([[1,2,3,4]])
array2 = np.array([[1],[2],[3],[4]])
print(array1.shape)
print(array2.shape)
```

```python
array1*array2
```

```python
array1 = np.array([[1,2,3,4],
                   [5,6,7,8]])
print(array1.shape)
print(array2.shape)
print(array1*array2)
```
