An important characteristic of the Python language is the consistency of its object model. Every number, string, data structure, function, class, module, and so on exists in the Python interpreter in its own “box,” which is referred to as a Python object. Each object has an associated type (e.g., integer, string, or function) and internal data. In practice this makes the language very flexible, as even functions can be treated like any other object. If you create object a, and make b = a, it won't create a copy; it will make them refer to the same object.

.copy() function can be used to create a shallow copy of an object.

```python
new_list = a.copy()
print(new_list)
new_list.append(6)
print(a)
print(new_list)
```

There is a difference between shallow copies and deep copies.

In shallow copies, nested objects are still shared. In deep copies, each object found within an object is recreated.

```python
a = [[1,2], [3,4], [5,6]]
shallow = a.copy()
print("Before change: ")
print(f"""Original list: {a}
Shallow copy: {shallow}""")
a[1][1] = 98
print("\nAfter changing only the original array:")
print(f"Original list: {a}")
print(f"Shallow copy: {shallow}")
```

The copy library can be used for creating deep copies.

```python
import copy as cp
deepcopy = cp.deepcopy(a)
print(deepcopy)
```

```python
deepcopy[0][0] = 46
print(f"the original list: {a}")
print(f"shallow copy with nested lists shared: {shallow}")
print(f"deep copy with completely separate nested lists: {deepcopy}")
```

## Control Flow

if statements are written with no parantheses, followed by colons.

Separate if blocks will evaluate each condition. But elif blocks will run only if the previous condition was incorrect.

If any of the conditions are True, no further elif or else blocks will be reached.

```python
x = 0
if x == 0:
  print("x is 0")
if x < 5:
  print("x is smaller than 5")
if x > 5:
  print("x is greater than 5")
```

```python
x = 5
if x == 0:
  print("x is 0")
elif x < 5:
  print("x is smaller than 5")
elif x > 5:
  print("x is greater than 5")
else:
  print("x is 5")
```

With and/or, conditions are evaluated left to right and will short-circuit.

the comparison c > d never gets evaluated because the first comparison was True.

```python
a = 5; b = 7

c = 8; d = 4

if a < b or c > d:
  print("one or both are correct")

if a < b and c > d:
  print("both are correct")

if d < a < b < c:
  print("correct")
```

## For loops

for loops are for iterating over a collection(like a list or a tuple) or an iterator. (An iterator in Python is an object that lets you traverse (loop through) elements of a collection — like lists, tuples, or strings — one at a time.)

**for** value **in** collection: do something with value

```python
my_list = ["hello", 5, 6.4, True, (5,6)]
for item in my_list:
  print(type(item))
```

for loops are basically iterators under the hood:

Iterable: An iterable is any object that can return an iterator using the iter() function.
Examples: lists, strings, tuples, dictionaries, sets…

```python
iterator = iter(my_list)   # get iterator from iterable
type(iterator)
```

An iterator is an object that keeps track of where it is during iteration and returns the next item when you call next().

```python
print(next(iterator))  # 1
print(next(iterator))  # 2
print(next(iterator))  # 3
print(next(iterator))  # 4
print(next(iterator))  # 5
#print(next(iterator)) raises StopIteration
```

***continue*** keyword: advance for loop to the next iteration, skipping the remainder of the block.

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
print(total)
```

***break*** keyword: exit the loop
if there are embedded loops, it exits the one where it is located

```python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
```

```python
for i in range(2):
    for j in range(3):
        if j == 1:
            break
        print(i, j)
```

## While loops

the loop will end when the condition evaluates to False or when it is exited with the break keyword

```python
x = 256
total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
```

As you can see, range produces integers up to but not including the endpoint. A common use of range is for iterating through sequences by index:



```python
seq = [1, 2, 3, 4]
print(len(seq))

for i in range(len(seq)):
   print(f"element {i}: {seq[i]}")
```

While you can use functions like list to store all the integers generated by range in some other data structure, often the default iterator form will be what you want. This snippet sums all numbers from 0 to 99,999 that are multiples of 3 or 5:

```python
total = 0

for i in range(100_000):
  # % is the modulo operator
  if i % 3 == 0 or i % 5 == 0:
    total += i

print(total)
```
