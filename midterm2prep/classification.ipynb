```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
df.head()
```

```python
df["diagnosis"] = df["diagnosis"].map({"B":0,"M":1})
df["diagnosis"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop("diagnosis",axis=1)
y = df["diagnosis"]
```

```python
X
```

```python
y
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=42,stratify=y)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_validate

# 1. Model ve Metriklerin Belirlenmesi
lda = LinearDiscriminantAnalysis()
scoring = ['accuracy', 'precision', 'recall', 'f1']

# 2. Cross-Validation Uygulama (cv=10 diyelim)
results = cross_validate(lda, X_train_scaled, y_train, cv=10, scoring=scoring)

# 3. Sonuçları bir DataFrame'e dönüştürme
# results sözlüğü 'test_accuracy', 'test_precision' gibi anahtarlar içerir
cv_df = pd.DataFrame(results)

# Sadece metrik sütunlarını alalım (fit_time vb. hariç)
metrics_df = cv_df[['test_accuracy', 'test_precision', 'test_recall', 'test_f1']]
metrics_df.columns = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

print("10-Fold CV Sonuçları:")
print(metrics_df)

# 4. Grafikleştirme (Boxplot - Her metriğin dağılımını görmek için)
plt.figure(figsize=(10, 6))
sns.boxplot(data=metrics_df)
plt.title('LDA Model Performans Dağılımı (10-Fold CV)')
plt.ylabel('Skor')
plt.show()

# 5. Alternatif: Her fold'u çizgi grafiği olarak görme
metrics_df.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Fold Bazlı Performans Değişimi')
plt.xlabel('Fold İndeksi')
plt.ylabel('Skor')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# LDA için 10-katlı çapraz doğrulama
lda_scores = cross_val_score(LinearDiscriminantAnalysis(), X_train_scaled, y_train, cv=10)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), lda_scores, marker='o', color='purple', linestyle='--')
plt.title("LDA 10-Fold CV Sonuçları")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
