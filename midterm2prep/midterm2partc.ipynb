Write a function
def e(N)
which calculates e according to the following equation.
Do not assume there is a factorial function defined.
e = 1 + 1
1! + 1
2! + . . . + 1
N !

```python
def e(N):
    e = 1.0
    factorial = 1
    for i in range (1,N+1):
        factorial *=i
        e += 1/factorial
    return e
```

Write a function
def containsK(a, K)
which checks if the array a contains two numbers whose
sum is K. Returns true if there are two such numbers,
false otherwise.

```python
def containsK(a,K):
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[i]+a[j]==K:
                return True
    return False
```
