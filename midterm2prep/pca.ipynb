```python
import learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab as p
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
X = df
print(X.shape)
print(X.head())
```

```python
print(y.shape)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,train_size=0.7,random_state=42,shuffle=True,stratify=y)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca_full = PCA()
pca_full.fit(X_train_scaled)
```

```python
print(pca_full.explained_variance_)
```

```python
pca_full.components_
```

```python
pca_full.explained_variance_ratio_
```

```python
explained_variance = pca_full.explained_variance_ratio_*100
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,5),np.cumsum(explained_variance),marker='o',linestyle='--')
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of components")
plt.ylabel('Cumulative Explained Variance in %')
plt.grid(alpha=0.3)
plt.show()
```

```python
pca_2d = PCA(n_components=2)
X_train_pca = pca_2d.fit_transform(X_train_scaled)

print(f"New X shape after PCA: {X_train_pca.shape} ")
```

```python
plt.figure(figsize=(8,5))
scatter = plt.scatter(X_train_pca[:,0], X_train_pca[:,1],c=y_train,cmap='ocean',edgecolor='k',s=60)
plt.title('PCA Projection of Iris dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Linear Discriminant Analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)

X_train_lda = lda.fit_transform(X_train_scaled,y_train)

plt.figure(figsize=(8,5))
scatter = plt.scatter(X_train_lda[:,0], X_train_lda[:,1],c=y_train,cmap='coolwarm',edgecolor='k',s=60)
plt.title('LDA of Iris Dataset')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred_lda = lda.predict(X_test_scaled)
print("--- LDA Classifier Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lda):.4f}")
print(classification_report(y_test, y_pred_lda))
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

lda_clf = LinearDiscriminantAnalysis()
lda_clf.fit(X_train_scaled,y_train)
y_pred_lda = lda_clf.predict(X_test_scaled)
acc_lda = accuracy_score(y_test,y_pred_lda)
```

```python
qda_clf = QuadraticDiscriminantAnalysis()
qda_clf.fit(X_train_scaled,y_train)
y_pred_qda = qda_clf.predict(X_test_scaled)
acc_qda = accuracy_score(y_test,y_pred_qda)
```

```python
print("--- Classifier Comparison ---")
print(f"LDA Accuracy: {acc_lda:.4f}")
print(f"QDA Accuracy: {acc_qda:.4f}\n")

print("QDA Detailed Report:")
print(classification_report(y_test, y_pred_qda))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Bin aralıklarını sabitlemek için tüm veri üzerinden hesapla (binlerin hizalı olması şart)
bins = 25
_, bin_edges = np.histogram(X_train_lda_bc, bins=bins)

# 2. Her sınıf için frekans (count) değerlerini aynı bin aralıklarıyla hesapla
hist0, _ = np.histogram(X_train_lda_bc[y_train_bc == 0], bins=bin_edges)
hist1, _ = np.histogram(X_train_lda_bc[y_train_bc == 1], bins=bin_edges)

# 3. Kesişen (overlap) kısmı bul: Her bin için iki değerden küçük olanı alıyoruz
intersection = np.minimum(hist0, hist1)

# --- Çizim Kısmı ---
plt.figure(figsize=(10,5))

# Orijinal histogramlar (yine senin renklerinle)
plt.hist(X_train_lda_bc[y_train_bc == 0], bins=bin_edges, alpha=0.7, color='crimson',
         label='Malignant (Class:1)', edgecolor='black')
plt.hist(X_train_lda_bc[y_train_bc == 1], bins=bin_edges, alpha=0.7, color='steelblue',
         label='Benign (Class:0)', edgecolor='black')

# Kesişen bölgeyi yeşil ile boya
# plt.bar kullanarak tam binlerin üzerine bindiriyoruz
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
plt.bar(bin_centers, intersection, width=np.diff(bin_edges), color='green',
        alpha=0.9, label='Overlap (Kesişim)', edgecolor='black')

# Diğer detaylar
plt.title("1D LDA Projection Of Breast Cancer Dataset")
plt.xlabel("Linear Discriminant 1")
plt.ylabel("Number of Patients")
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
