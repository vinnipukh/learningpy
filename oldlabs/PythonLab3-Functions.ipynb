## **Functions in Python**

In data science, it is good practice to define functions for operations we perform regularly.

We define functions using the **def** keyword and colons. The function block is indented. The function can either return something using the **return** keyword or only have a side effect like printing something.

```python
def add_numbers(a,b,c):
  return a+b+c

total = add_numbers(2,3,4)
total
```

```python
def print_string(x):
  print(x) # prints the variable and returns None

print_string("hi there")
```

Functions have positional arguments and keyword arguments. As their name suggests, positional arguments should be specified according to their positions among the parameters. They must be specified while calling the function.

Keyword arguments should be specified using the keywordand the equal sign. They usually have default values specified in the function definition. They are usually optional.

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Joseph")
greet("Joseph", greeting="Goodbye")
greet("Joseph", "Goodbye")
greet(name="Joseph", greeting="Good Morning")
# greet(name="Joseph", "Good Morning") you can't do this
```

```python
def show(*args):
    print(args)

show("hello",[1, 2], True, 4)
```

```python
def show(**kwargs):
    print(kwargs)

show(name="Daenerys", surname="Targaryen", age=None)
```

Functions can return multiple values or objects.

```python
def analyze_numbers(nums):
    total = sum(nums)
    avg = total / len(nums)
    return total, avg, len(nums)

t, a, n = analyze_numbers([2, 4, 6, 8])
print(f"Sum: {t}, Average: {a}, Count: {n}")
```

Sometimes it might be preferable and more readable to return a dictionary instead of a tuple.

```python
def student_stats(name, grades):
    avg = sum(grades) / len(grades)
    highest = max(grades)
    return {"name": name, "average": avg, "highest": highest}

info = student_stats("Ayşe", [90, 85, 100])
print(info["average"])
```

### Functions as Objects

In Python, functions are first-class objects.
That means functions behave like any other value. You can:
- assign them to variables,
- pass them as arguments to other functions,
- return them from functions, and
- store them in data structures (lists, dicts, etc.)

```python
# Assigning functions to new variables
def greet(name):
    return f"Hello, {name}!"
# say_hello and greet both point to the same function object in memory.
# no paranthesis means your are not calling the function, just referencing it
say_hello = greet
print(say_hello("Jon Snow"))
```

```python
def shout(text):
    return text.upper()

def whisper(text):
    return text.lower()

def speak(func, message): # you are passing a function to this function
    return func(message) # you call the function inside this function

print(speak(shout, "Hello"))
print(speak(whisper, "Hello"))
```

```python
# you can create a function within a function and return it.
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

times3 = make_multiplier(3)
times5 = make_multiplier(5)
# you can then use the returned function with parameters
print(times3(10))
print(times5(10))
```

```python
# you can create functions on one line
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
# you can store functions inside data structures like dicts
operations = {
    "add": add,
    "subtract": subtract,
    "multiply": multiply
}
# you can efficiently access whichever function you need with the key
print(operations["add"](2, 3))
print(operations["multiply"](4, 5))
```

### Lambda Functions

```python
def square(x):
    return x ** 2
square(2)
```

```python
square = lambda x: x ** 2
print(square(5))
```

```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

The map() function applies a given function to every item in an iterable (like a list, tuple, etc.)
and returns a new map object (which can be converted into a list, tuple, etc.).

map() doesn’t give you a list automatically: it gives you a map object, which is an iterator.

You usually convert it to a list using list() for display.

```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]

result = map(square, numbers)
# “Take this function and map it onto every element of this list.”

print(result)          # Output: <map object at ...>
print(list(result))    # Output: [1, 4, 9, 16, 25]
```

### Generators

We have mentioned iterators in the first lab. Generators work similarly, but instead of iterating through objects, they create/generate iterable objects.

A generator is a special kind of function that remembers its state and yields items one at a time instead of returning them all at once.
They are used to create iterators efficiently — especially for large data.

```python
def normal_function():
    return [1, 2, 3]

def generator_function():
    yield 1
    yield 2
    yield 3

print(normal_function())       # Output: [1, 2, 3]
print(generator_function())    # Output: <generator object ...>
```

```python
for value in generator_function():
    print(value)
```

Normally, a function that returns a list builds the entire list in memory before returning it.
Generators, on the other hand:
- Produce values one at a time (lazy evaluation)
- Don’t store all results in memory
- Are perfect for processing large datasets or infinite sequences

yield is what makes a function a generator.
Each time Python hits yield, it pauses the function — remembers where it left off —
and resumes from that point on the next iteration.

```python
def _make_gen():
    for x in range(100):
        yield x ** 2
gen = _make_gen()
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
import sys
nums = [x**2 for x in range(10000)] # a regular list created with list comprehension
gen = (x**2 for x in range(10000)) # a generator using generator expression (parantheses)
print(type(nums))
print(type(gen))
print(sys.getsizeof(nums))  # much larger
print(sys.getsizeof(gen))   # much smaller
```
