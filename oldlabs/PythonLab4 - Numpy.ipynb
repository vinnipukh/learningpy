Creating vectors: 1D

```python
data = np.array([1.5, -0.1, 3])
print(data)
print(data.shape)
```

```python
data3 = np.array([[[1.5, -0.1, 3],
                  [0, -3, 6.5]]])
print(data3)
print(data3.shape) # 1 block, 2 rows, 3 columns.
print(data3.ndim)
```

```python
data4 = np.array([
    [[1.5, -0.1, 3],
     [0, -3, 6.5]],

    [[2.1, 0.5, -1],
     [4.2, 3.3, 0]]
])
print(data4)
print(data4.shape) # 2 blocks, 2 rows, 3 columns.
print(data4.ndim)
```

convert a multi-dimensional array into a 1D array — but they differ in whether they create a copy or a view of the data.

```python
arr = np.arange(15).reshape((5, 3))
print(arr)
raveled = arr.ravel()
raveled
```

```python
raveled[5] = 6565
arr
```

```python
flattened = arr.flatten()
print(flattened)
flattened[5] = 7777
print(arr)
```

```python
arr_transposed = arr.T
arr_transposed
```

```python
arr = np.array([[0, 1, 0],
                [1, 2, -2],
                [6, 3, 2],
                [-1, 0, -1],
                [1, 0, 1]])
print(arr.T)
print(arr)
# inner matrix product
np.dot(arr, arr.T)
```

Concatenating, stacking and splitting

```python
print(x.shape)
intercept_value = np.ones((x.shape[0], 1))
print(intercept_value)
print(intercept_value.shape)
```

```python
x_with_intercept = np.concatenate((intercept_value, x), axis=1)
x_with_intercept
```

```python
x_with_intercept2 = np.concatenate((intercept_value, x), axis=0) # add rows
x_with_intercept2
```

The arrays must have the same shape along all other axes.

For axis=0, they must have the same number of columns.

For axis=1, they must have the same number of rows.

```python
a1 = np.array([[1,1],
               [2,2]])
a2 = np.array([[3,3],
              [4,4]])
print(a1)
print(a2)
```

```python
# Stack arrays vertically
vert_stack = np.vstack((a1, a2))
print(vert_stack)
```

```python
# Stack arrays horizontally
horz_stack = np.hstack((a1, a2))
print(horz_stack)
```

### **Linear Algebra**

"*" = element-wise product

dot() or @ infix operator = matrix multiplication

```python
np.random.seed(42)
x = np.random.randn(2, 3) # 2x3
y = np.random.rand(2 , 3)
print(f"x = \n{x}")
print(f"y = \n{y}")
```

```python
x.dot(y.T)
```

```python
np.dot(x, y.T)
```

```python
x @ y.T
```

```python
vector1 = np.random.rand(4,)
vector2 = np.random.rand(4,)
print(vector1.shape)
```

```python
np.dot(vector1, vector2)
```

```python
vector1 @ vector2
```

```python
C = np.random.rand(2, 3, 4)
D = np.random.rand(2, 4, 5)
print("\nTensor product shape:", np.matmul(C, D).shape)
# (2, 3, 5) -> Batch matrix multiplication
```

```python
square_matrix = np.random.randn(4,4)
print(square_matrix)
inverse = np.linalg.inv(square_matrix)
inverse
```

```python
np.linalg.eig(inverse)
```
