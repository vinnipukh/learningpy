Map Ordinals: Convert Ex/Gd/TA to 5/4/3 integers.

Dummies: Convert Neighborhood, Zone, and MSSubClass to binary columns.

Find all rows where Garage Yr Blt equals 2207, and replace that value with 2007.

1. Create a boolean mask

3. Assign a corrected value

```python
# Fix the specific typo found in value_counts (2207 -> 2007)
df.loc[df['Garage Yr Blt'] == 2207, 'Garage Yr Blt'] = 2007
```

Pandas .map() applies a function to each element in a Series, i.e., each cell in the column.

map.get(x,x)

If key exists in the dictionary → return its mapped value

If key does NOT exist → return the fallback (default)

```python
# --- 2. ORDINAL ENCODING (Preserving Rank) ---
# Map 1: Standard Quality Columns (Ex, Gd, TA...)
quality_map = {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}
# These are rankings (excellent > good > typical > fair > poor
qual_cols = ['Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 'Heating QC',
             'Kitchen Qual', 'Garage Qual', 'Garage Cond']

# Use 'get' to handle cases where a specific category might not exist in a column
for col in qual_cols:
    df[col] = df[col].map(lambda x: quality_map.get(x, x))

# Map 2: Basement Exposure (None -> No -> Mn -> Av -> Gd)
bsmt_exposure_map = {'None': 0, 'No': 1, 'Mn': 2, 'Av': 3, 'Gd': 4}
df['Bsmt Exposure'] = df['Bsmt Exposure'].map(lambda x: bsmt_exposure_map.get(x, x))

# Map 3: Lot Shape (Regular -> Irregular)
lot_shape_map = {'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3}
df['Lot Shape'] = df['Lot Shape'].map(lambda x: lot_shape_map.get(x, x))

# Map 4: Land Slope (Gentle -> Severe)
slope_map = {'Gtl': 0, 'Mod': 1, 'Sev': 2}
df['Land Slope'] = df['Land Slope'].map(lambda x: slope_map.get(x, x))
```

**get_dummies()**

1. Detects all categorical columns (object or category dtype)

2. For each column, identifies all unique categories

3. Creates a new binary column for each category

4. Fills with 0/1 depending on the row’s category

5. Drops the original column

```python
# --- 3. NOMINAL ENCODING (One-Hot) ---
# Convert MSSubClass to string because 20, 30, 40 are codes, not math
df['MS SubClass'] = df['MS SubClass'].astype(str)

# Separate Target Variable before encoding/scaling
if 'SalePrice' in df.columns:
    X = df.drop('SalePrice', axis=1)
    y = df['SalePrice']
else:
    X = df.copy()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# 3. DEFINE THE TRANSFORMER (Encoding + Scaling)
# ==========================================
# Identify categorical (nominal) vs numerical columns automatically
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = X.select_dtypes(include=['number']).columns.tolist()

print(f"Categorical Columns to Encode: {len(categorical_cols)}")
print(f"Numerical Columns to Scale: {len(numerical_cols)}")

# Create the ColumnTransformer
# - OneHotEncoder: Converts categories to binary. handle_unknown='ignore' prevents crashes on new data.
# - StandardScaler: Scales numbers to mean=0, std=1.
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), categorical_cols)
    ],
    verbose_feature_names_out=False # Keeps column names clean (e.g., "Neighborhood_OldTown" instead of "cat__Neighborhood_OldTown")
)

# Apply the transformations
# This replaces both get_dummies AND the separate scaler step
X_processed = preprocessor.fit_transform(X)

# Convert back to DataFrame for readability
# We need to extract the new feature names generated by OneHotEncoder
feature_names = preprocessor.get_feature_names_out()
X_processed_df = pd.DataFrame(X_processed, columns=feature_names)

print("Preprocessing with sklearn OneHotEncoder Complete.")
display(X_processed_df.head())
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# 1. Run PCA without specifying n_components (it keeps them all by default)
pca = PCA()
pca.fit(X_processed_df)

# 2. Calculate variance ratios
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# 3. Plotting
plt.figure(figsize=(12, 6))

# Bar chart for individual components
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual variance')

# Line chart for cumulative variance
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', color='red', label='Cumulative variance')

# Visual guide lines
plt.axhline(y=0.90, color='green', linestyle='--', label='90% Explained Variance')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component Index')
plt.title('Scree Plot')
plt.legend(loc='best')
plt.grid(True)

# Limit x-axis to first 50 components for readability (since the tail is long and flat)
plt.xlim(0, 50)
plt.show()

# Print exact number of components needed for 90% variance
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
print(f"Number of components needed to explain 90% of variance: {n_components_90}")
```

## Regression Analysis

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# --- EXPERIMENT 1: BASELINE (All Features) ---
# Split the original scaled data
X_train_all, X_test_all, y_train, y_test = train_test_split(X_processed_df, y, test_size=0.2, random_state=42)

# Train
lr_all = LinearRegression()
lr_all.fit(X_train_all, y_train)

# Predict
y_pred_all = lr_all.predict(X_test_all)

# --- EXPERIMENT 2: PCA TRANSFORMED (125 Components) ---
# Apply PCA
pca = PCA(n_components=125)
X_pca = pca.fit_transform(X_processed_df)

# Split the PCA data (MUST use same random_state to be fair!)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Train
lr_pca = LinearRegression()
lr_pca.fit(X_train_pca, y_train_pca)

# Predict
y_pred_pca = lr_pca.predict(X_test_pca)

# --- RESULTS COMPARISON ---
def print_metrics(name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return [mae, rmse, r2]

metrics_all = print_metrics("Baseline (All Features)", y_test, y_pred_all)
metrics_pca = print_metrics("PCA (125 Components)", y_test, y_pred_pca)

# Create a nice DataFrame to compare
results_df = pd.DataFrame({
    'Metric': ['MAE (Lower is Better)', 'RMSE (Lower is Better)', 'R2 Score (Higher is Better)'],
    'Baseline (260 Features)': metrics_all,
    'PCA (125 Components)': metrics_pca
})

# Calculate the difference
results_df['Difference'] = results_df['PCA (125 Components)'] - results_df['Baseline (260 Features)']

display(results_df.round(4))
```

We threw away 10% of the information (variance), so naturally, the model knows slightly less. However, we achieved nearly the same result with half the complexity. In a production environment, the simpler model is often preferred for speed and stability.

This means the first version was not overfitting.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

# --- Plot 1: Baseline (All Features) ---
plt.subplot(1, 2, 1)
sns.scatterplot(x=y_test, y=y_pred_all, alpha=0.5, color='blue')
# Draw a red line representing perfect prediction
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title(f"Baseline (260 Features)\nR2: {metrics_all[2]:.4f}")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid(True)

# --- Plot 2: PCA (125 Components) ---
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_test, y=y_pred_pca, alpha=0.5, color='green')
# Draw a red line representing perfect prediction
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title(f"PCA (125 Components)\nR2: {metrics_pca[2]:.4f}")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid(True)

plt.tight_layout()
plt.show()
```

the "dots" look almost identical, proving PCA worked.

# Ames Housing Dataset: Data Dictionary

### 1. Building Basics (Type, Age, Style)
* **`MSSubClass`**: The building class (e.g., 20=1-Story 1946+, 60=2-Story 1946+). *Note: These look like numbers but are categorical codes.*
* **`MSZoning`**: The general zoning classification (e.g., RL=Residential Low Density, RM=Residential Medium Density).
* **`BldgType`**: Type of dwelling (e.g., 1Fam=Single-family detached, TwnhsE=Townhouse End).
* **`HouseStyle`**: Style of dwelling (e.g., 1Story, 2Story, SFoyer=Split Foyer).
* **`YearBuilt`**: Original construction date.
* **`YearRemodAdd`**: Remodel date (same as construction date if no remodeling or additions).

### 2. Space & Area (Quantitative)
* **`LotFrontage`**: Linear feet of street connected to property.
* **`LotArea`**: Lot size in square feet.
* **`1stFlrSF`**: First Floor square feet.
* **`2ndFlrSF`**: Second Floor square feet.
* **`LowQualFinSF`**: Low quality finished square feet (all floors).
* **`GrLivArea`**: Above grade (ground) living area square feet.
* **`TotRmsAbvGrd`**: Total rooms above grade (does not include bathrooms).

### 3. Quality & Condition
* **`OverallQual`**: Overall material and finish quality (Scale 1-10). **(Crucial Predictor)**
* **`OverallCond`**: Overall condition rating (Scale 1-10).
* **`Functional`**: Home functionality rating (e.g., Typ=Typical, Min1=Minor Deductions).

### 4. Basement Features
* **`Foundation`**: Type of foundation (e.g., PConc=Poured Concrete, CBlock=Cinder Block).
* **`BsmtQual`**: Height of the basement (evaluates quality). *NaN = No Basement*.
* **`BsmtCond`**: General condition of the basement. *NaN = No Basement*.
* **`BsmtExposure`**: Refers to walkout or garden level basement walls (e.g., Gd=Good, No=No Exposure).
* **`BsmtFinType1`**: Quality of basement finished area (e.g., GLQ=Good Living Quarters).
* **`BsmtFinSF1`**: Type 1 finished square feet.
* **`BsmtFinType2`**: Quality of second finished area (if present).
* **`BsmtFinSF2`**: Type 2 finished square feet.
* **`BsmtUnfSF`**: Unfinished square feet of basement area.
* **`TotalBsmtSF`**: Total square feet of basement area.

### 5. Garage Features
* **`GarageType`**: Garage location (e.g., Attchd=Attached, Detchd=Detached). *NaN = No Garage*.
* **`GarageYrBlt`**: Year garage was built.
* **`GarageFinish`**: Interior finish of the garage (Fin=Finished, RFn=Rough Finished).
* **`GarageCars`**: Size of garage in car capacity.
* **`GarageArea`**: Size of garage in square feet.
* **`GarageQual`**: Garage quality.
* **`GarageCond`**: Garage condition.

### 6. Location & Lot Configuration
* **`Street`**: Type of road access (Pave, Grvl).
* **`Alley`**: Type of alley access. *NaN = No Alley access*.
* **`LotShape`**: General shape of property (Reg=Regular, IR1=Slightly Irregular).
* **`LandContour`**: Flatness of the property (Lvl=Level, Bnk=Banked).
* **`LotConfig`**: Lot configuration (Inside, Corner, CulDSac).
* **`LandSlope`**: Slope of property (Gtl=Gentle, Mod=Moderate).
* **`Neighborhood`**: Physical locations within Ames city limits.
* **`Condition1`**: Proximity to main road or railroad (e.g., Norm=Normal, RRAn=Adjacent to Railroad).
* **`Condition2`**: Proximity to main road or railroad (if a second is present).

### 7. Utilities & Comfort
* **`Utilities`**: Type of utilities available (AllPub=All public, NoSeWa=Electricity & Gas only).
* **`Heating`**: Type of heating (GasA=Gas forced air, Grav=Gravity).
* **`HeatingQC`**: Heating quality and condition.
* **`CentralAir`**: Central air conditioning (Y/N).
* **`Electrical`**: Electrical system (SBrkr=Standard Circuit Breakers, FuseA=Fuse Box).
* **`Fireplaces`**: Number of fireplaces.
* **`FireplaceQu`**: Fireplace quality. *NaN = No Fireplace*.

### 8. Bathrooms
* **`BsmtFullBath`**: Basement full bathrooms.
* **`BsmtHalfBath`**: Basement half bathrooms.
* **`FullBath`**: Full bathrooms above grade.
* **`HalfBath`**: Half baths above grade.

### 9. Exterior Features (Roof & Siding)
* **`RoofStyle`**: Type of roof (Gable, Hip).
* **`RoofMatl`**: Roof material (Shingle, Slate).
* **`Exterior1st`**: Exterior covering on house (VinylSd, MetalSd, BrkFace).
* **`Exterior2nd`**: Secondary exterior covering.
* **`MasVnrType`**: Masonry veneer type (BrkFace, Stone).
* **`MasVnrArea`**: Masonry veneer area in square feet.
* **`ExterQual`**: Exterior material quality.
* **`ExterCond`**: Present condition of the material on the exterior.

### 10. Outdoor & Misc
* **`PavedDrive`**: Paved driveway (Y=Paved, N=Dirt/Gravel, P=Partial).
* **`WoodDeckSF`**: Wood deck area in sq ft.
* **`OpenPorchSF`:** Open porch area in sq ft.
* **`EnclosedPorch`**: Enclosed porch area in sq ft.
* **`3SsnPorch`**: Three season porch area in sq ft.
* **`ScreenPorch`**: Screen porch area in sq ft.
* **`PoolArea`**: Pool area in sq ft.
* **`PoolQC`**: Pool quality. *NaN = No Pool*.
* **`Fence`**: Fence quality. *NaN = No Fence*.
* **`MiscFeature`**: Miscellaneous feature (Elevator, Tennis Court).
* **`MiscVal`**: $Value of miscellaneous feature.

### 11. Sale Information
* **`MoSold`**: Month Sold.
* **`YrSold`**: Year Sold.
* **`SaleType`**: Type of sale (WD=Warranty Deed, New=New Construction).
* **`SaleCondition`**: Condition of sale (Normal, Abnorml, Partial).
* **`SalePrice`**: **(Target Variable)** The property's sale price in dollars.
